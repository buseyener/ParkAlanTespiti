```python
import cv2
import pickle
import numpy as np 

# Video kaynağını aç
cap = cv2.VideoCapture("video.mp4")

# 'noktalar' dosyasını yükle
try:
    with open("noktalar", "rb") as f:
        liste = pickle.load(f)
except FileNotFoundError:
    print("Dosya bulunamadı! Liste boş oluşturulacak.")
    liste = []

def check(frame1):
    spacecounter = 0
    for pos in liste:
        x, y = pos

        # Dikdörtgenin boyutları
        width = 30  # Uzun kenar
        height = 11  # Kısa kenar
        
        # Kontrol edilecek alanı kes
        crop = frame1[y:y + height, x:x + width]
        count = cv2.countNonZero(crop)
        
        # Kontrol edilen alanın durumuna göre renk belirle
        if count < 150:
            color = (0, 255, 0)  # Boş alan
            spacecounter += 1
        else:
            color = (0, 0, 255)  # Dolu alan

        # Her bir dikdörtgeni çiz
        cv2.rectangle(frame1, (x, y), (x + width, y + height), color, 2)

    # Boş alan sayısını ekrana yazdır
    cv2.putText(frame1, f"bos: {spacecounter}/{len(liste)}", (15, 24), cv2.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0), 4)

while True:
    ret, frame = cap.read()
    if not ret:  # Video bittiğinde döngüyü kır
        print("Video sona erdi veya okunamadı.")
        break

    gri = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blur = cv2.GaussianBlur(gri, (3, 3), 1)
    thres = cv2.adaptiveThreshold(blur, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 25, 16)
    median = cv2.medianBlur(thres, 5)
    dilates = cv2.dilate(median, np.ones((3, 3)), iterations=1)

    check(dilates)  # Kontrol fonksiyonunu çağır
    cv2.imshow("asd", frame)  # Çerçeveyi göster

    if cv2.waitKey(200) & 0xFF == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
